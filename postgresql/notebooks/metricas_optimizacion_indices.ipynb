```python
import pandas as pd
import matplotlib.pyplot as plt

data = [
    {
        "query": "Historial órdenes por cliente",
        "execution_before_ms": 15.442,
        "execution_after_ms": 9.210,
        "improvement_percent": 40.36
    },
    {
        "query": "Ventas por categoría y mes",
        "execution_before_ms": 1902.142,
        "execution_after_ms": 395.193,
        "improvement_percent": 79.22
    },
    {
        "query": "Desempeño vendedores por región",
        "execution_before_ms": 424.708,
        "execution_after_ms": 397.821,
        "improvement_percent": 6.33
    },
    {
        "query": "Productos con reseñas negativas",
        "execution_before_ms": 912.709,
        "execution_after_ms": 2408.446,
        "improvement_percent": -163.87
    }
]

df = pd.DataFrame(data)

plt.figure(figsize=(11, 6))
plt.bar(df["query"], df["improvement_percent"])
plt.xticks(rotation=35, ha="right")
plt.ylabel("Mejora (%)")
plt.title("Mejora de rendimiento PostgreSQL después de índices")
plt.axhline(0)
plt.tight_layout()
plt.savefig("grafica_mejora_postgresql.png", dpi=150)
plt.show()
```
